```python
(x_train, y_train), (x_test, y_test) = boston_housing.load_data()
```

```python
x_train.shape
```

```python
y_train.shape
```

```python
x_test.shape, y_test.shape
```

```python
from keras.models import Sequential
from keras.layers import Dense
from tensorflow.keras import optimizers
```

```python
model = Sequential()
```

```python
#Hidden Layer - 1
model.add(Dense(1000, input_shape = (x_train.shape[1],), activation='sigmoid'))
#Hidden Layer - 2
model.add(Dense(1000, activation = 'sigmoid'))
#Hidden Layer - 3
model.add(Dense(1000, activation = 'sigmoid'))
# Output Layer
model.add(Dense(1))
```

```python
sgd = optimizers.SGD()
```

```python
model.compile(optimizer = sgd, loss = 'mean_squared_error', metrics= ['MeanSquaredError'] )
```

```python
model.summary()
```

```python
model.fit(x_train, y_train, epochs = 1000, batch_size=10, verbose=1)
```

```python
model.evaluate(x_train,y_train)
```

```python
model.evaluate(x_test,y_test)
```

```python
test_pred = model.predict(x_test)
```
