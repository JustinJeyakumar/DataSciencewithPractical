```python
from sklearn.pipeline import Pipeline, make_pipeline
from sklearn.datasets import load_iris
import numpy as np
import pandas as pd
from sklearn.model_selection import train_test_split

from sklearn.preprocessing import StandardScaler
from sklearn.impute import SimpleImputer
from sklearn.preprocessing import LabelEncoder

from sklearn.ensemble import RandomForestClassifier
from sklearn.model_selection import GridSearchCV, RandomizedSearchCV
import joblib
```

```python
data = load_iris()
```

```python
X = data['data']
#X = pd.DataFrame(X,columns = data['feature_names'])
```

```python
X.shape
```

```python
flower = ['rose','jasmine','lily']
```

```python
new_col = []
for i in range(0,50):
    new_col.append(flower)
```

```python
X['flower'] = [xs for x in new_col for xs in x]
```

```python
y = data['target']
```

```python
X
```

```python
y
```

```python
train_X,test_X,train_y,test_y = train_test_split(X,y,test_size = 0.3, random_state = 12,stratify= y)
```

```python
train_X
```

```python
scale = StandardScaler()
```

```python
scale.fit(train_X)
```

```python
scale.fit_transform(train_X)
```

```python
scale.fit_transform(test_X)
```

```python
joblib.dump(scale, "scaling.pkl")
```

```python
train_X.iloc[0,1] = np.NaN
test_X.iloc[0,1] = np.NaN
```

```python
train_X.isna().sum()
```

```python
test_X.isna().sum()
```

```python
impute = SimpleImputer(strategy='mean')
```

```python
impute.fit_transform(train_X)
```

```python
joblib.dump(impute, 'impute.pkl')
```

```python
encode  = LabelEncoder()
```

```python
encode.fit_transform(train_X['flower'])
```

```python
joblib.dump(encode, 'encode.pkl')
```

```python
train_X
```

```python
pipe = Pipeline([('impute', SimpleImputer(strategy='mean')),
                 ('scale',StandardScaler()), 
                  ('model', RandomForestClassifier())])
```

```python
pipe.fit(train_X,train_y)
```

```python
pipe.predict(train_X)
```

```python
pipe.predict(test_X)
```

```python
joblib.dump(pipe, 'pipe.pkl')
```

```python
pipe = make_pipeline((SimpleImputer(strategy='mean')),
                 (StandardScaler()), 
                  (RandomForestClassifier()))
```

```python
parameters = [{
    'randomforestclassifier': [RandomForestClassifier()],
    'randomforestclassifier__n_estimators':[25,50,75,100], 
    'randomforestclassifier__max_depth':[1,2,3,4,5,6,7,8,9], 
    'randomforestclassifier__min_samples_leaf':[1,2,3,4,5]}]
```

```python
gridsearch = GridSearchCV(pipe,parameters,cv = 5, verbose = 1)
```

```python
gridsearch.fit(train_X,train_y)
```

```python
gridsearch.best_estimator_
```

```python
gridsearch.score(test_X, test_y)
```

```python
joblib.dump(gridsearch,"best_model_pipe.pkl")
```
