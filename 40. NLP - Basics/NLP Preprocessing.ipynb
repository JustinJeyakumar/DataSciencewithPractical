```python
def preprocessing(text):
    #text parsing from html
    soup = BeautifulSoup(text, "html.parser")
    text = soup.get_text()
    
    
    
    #accented character removal
    unaccented_string = unidecode.unidecode(accented_string)
    text = unidecode.unidecode(text)
    
    
    #contraction fix
    text = contractions.fix(text)
    
    #special character removal
    text = re.sub('[^A-Za-z0-9]+', ' ', text)
    
    
    #lower case convertion
    text = text.lower()
    
    
    #remove stopwords
    stopwords = nltk.corpus.stopwords.words('english')
    stopwords.remove('not')
    text_list = [word for word in text.split(' ') if word not in stopwords ]
    text = ' '.join(text_list)
    
    
    lemmatizer = WordNetLemmatizer()
    words = word_tokenize(text)
    lemma = []
    for w in words:
        lemma.append(lemmatizer.lemmatize(w, pos ="a"))

    text = ' '.join(lemma)
    
    return text
```

```python
preprocessed= []
for doc in Corpus:
    preprocessed.append(preprocessing(doc))
```

```python
from sklearn.feature_extraction.text import CountVectorizer
import pandas as pd
vectorizer = CountVectorizer(binary = True,ngram_range = (1,1))
```

```python
X = vectorizer.fit_transform(preprocessed)
df_bow_sklearn = pd.DataFrame(X.toarray(),columns=vectorizer.get_feature_names())
df_bow_sklearn.head()



print(vectorizer.get_feature_names())
```
